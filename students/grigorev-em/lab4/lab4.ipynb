```python
from source.pca import PCA
import matplotlib.pyplot as plt
```

```python
import sklearn.datasets as ds
df = ds.load_breast_cancer()
x = df['data']
y = df['target']
```

```python
cpca = PCA(n_components=3).fit(x)
```

```python
import numpy as np
```

```python
E = np.cumsum(cpca.lambdas)[::-1] / np.sum(cpca.lambdas)
dE = E[1:] - E[:-1]  
optim = 26
plt.plot(E, marker='*', label="Em")
plt.axline((optim, min(E)), (optim, 1), color='r', label="Оптимальное значение компонент")
plt.xlabel("Число компонент")
plt.ylabel("Em")
plt.grid()

plt.legend()
```

```python
from sklearn.decomposition import PCA as PCA_t
```

```python
x_t = PCA_t(n_components=2).fit_transform(x)
x_s = PCA(n_components=2).fit_transform(x)
```

```python
plt.figure(figsize=(12, 6))
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(18, 6))

ax[0].scatter(x_t[:, 0], x_t[:, 1], c=y)
ax[1].scatter(-x_s[:, 0], x_s[:, 1], c=y)
ax[0].set_title("PCA")
ax[1].set_title("Hand PCA")
```
